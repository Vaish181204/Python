```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#Load dataset
df_ghi = pd.read_csv('/content/GHI_Report.csv')
```

```python
df_ghi.head()
```

```python
df_ghi.tail()
```

```python
#Define features (Economy, Family, Health, Freedom) & target (Happiness Score)
X = df_ghi[['Economy', 'Fam', 'Health', 'Freedom']]
y = df_ghi['H_Score']
```

```python
#Split data : 3:1 ratio means 75% train, 25% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=1)
```

```python
#Create & train model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#Make predictions on test set
y_pred = model.predict(X_test)
```

```python
#Calculate RMSE
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
```

```python
print(f"The RMSE of the model is : {rmse}")
```
