```python
import math
import matplotlib.pyplot as plt
```

```python
points = {
    "P1": (0.1,0.6),
    "P2": (0.15,0.71),
    "P3": (0.08,0.9),
    "P4": (0.16,0.85),
    "P5": (0.2,0.3),
    "P6": (0.25,0.5),
    "P7": (0.24,0.1),
    "P8": (0.3,0.2)
}
```

```python
m1 = points["P1"]
m2 = points["P8"]

print("Initial m1:",m1)
print("Initial m2:",m2)
```

```python
def euclidean_distance(p1,p2):
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)
```

```python
cluster1 = []
cluster2 = []

for name, point in points.items():
    d1 = euclidean_distance(m1,point)
    d2 = euclidean_distance(m2,point)

    if d1 < d2:
      cluster1.append(name)
    else:
      cluster2.append(name)

print("Cluster 1:",cluster1)
print("Cluster 2:",cluster2)
```

```python
if "P6" in cluster1:
  print("P6 belongs to cluster 1")
else:
  print("P6 belongs to cluster 2")
```

```python
print("Poplution of cluster around m2:", len(cluster2))
```

```python
def update_centroid(cluster):
  x_sum = 0
  y_sum = 0

  for p in cluster:
    x_sum += points[p][0]
    y_sum += points[p][1]
  return (x_sum/len(cluster), y_sum/len(cluster))
```

```python
new_m1 = update_centroid(cluster1)
new_m2 = update_centroid(cluster2)

print("Updated m1:",new_m1)
print("Updated m2:",new_m2)
```

```python
#Cluster 1
x1 = [points[p][0] for p in cluster1]
y1 = [points[p][1] for p in cluster1]

#Cluster 2
x2 = [points[p][0] for p in cluster2]
y2 = [points[p][1] for p in cluster2]
```

```python
plt.scatter(x1,y1,color="red")
plt.scatter(x2,y2,color="blue")
```

```python
plt.figure()

#Plot clusters
plt.scatter(x1,y1)
plt.scatter(x2,y2)

#Initial centroids
plt.scatter(m1[0], m1[1], marker='X', s=150)
plt.scatter(m2[0], m2[1], marker='X', s=150)

#Updated cetroids
plt.scatter(new_m1[0], new_m1[1], marker='X', s=200)
plt.scatter(new_m2[0], new_m2[1], marker='X', s=200)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("K-Means Clustering (K=2)")

plt.show()
```

## Re-assign Points to Clusters (Second Iteration)

### Subtask:
Re-assign each point to cluster1 or cluster2 based on its Euclidean distance to the updated centroids (`new_m1` and `new_m2`) from the previous iteration. This is done to refine the cluster assignments as the centroids have moved, which is a core step in the K-Means iterative process. Iterate through all points, calculate distances to the current centroids, and update the cluster lists accordingly.


**Reasoning**:
The subtask requires re-assigning points to clusters based on the updated centroids (`new_m1` and `new_m2`). This involves initializing new cluster lists, iterating through each point, calculating its distance to both new centroids using the `euclidean_distance` function, and then assigning the point to the closest cluster. Finally, the new cluster assignments are printed.



```python
cluster1_iter2 = []
cluster2_iter2 = []

for name, point in points.items():
    d1_new = euclidean_distance(new_m1, point)
    d2_new = euclidean_distance(new_m2, point)

    if d1_new < d2_new:
        cluster1_iter2.append(name)
    else:
        cluster2_iter2.append(name)

print("Cluster 1 (Iteration 2):", cluster1_iter2)
print("Cluster 2 (Iteration 2):", cluster2_iter2)
```

**Reasoning**:
After re-assigning the points to clusters in the second iteration, the next step is to update the centroids based on these new cluster assignments. This will involve using the `update_centroid` function with `cluster1_iter2` and `cluster2_iter2` to get the second iteration's centroids.



```python
new_m1_iter2 = update_centroid(cluster1_iter2)
new_m2_iter2 = update_centroid(cluster2_iter2)

print("Updated m1 (Iteration 2):", new_m1_iter2)
print("Updated m2 (Iteration 2):", new_m2_iter2)
```

**Reasoning**:
To prepare for plotting the second iteration of clusters, I need to extract the x and y coordinates for each point in `cluster1_iter2` and `cluster2_iter2` into separate lists.



```python
x1_iter2 = [points[p][0] for p in cluster1_iter2]
y1_iter2 = [points[p][1] for p in cluster1_iter2]

x2_iter2 = [points[p][0] for p in cluster2_iter2]
y2_iter2 = [points[p][1] for p in cluster2_iter2]
```

**Reasoning**:
Having prepared the coordinate data for the second iteration clusters, the next step is to visualize these clusters along with all centroids (initial, first iteration, and second iteration) on a single plot to observe the K-Means progression.



```python
plt.figure(figsize=(8, 6))

# Plot clusters for Iteration 2
plt.scatter(x1_iter2, y1_iter2, color='red', label='Cluster 1 (Iter 2)')
plt.scatter(x2_iter2, y2_iter2, color='blue', label='Cluster 2 (Iter 2)')

# Plot initial centroids
plt.scatter(m1[0], m1[1], marker='X', s=150, color='black', label='Initial m1')
plt.scatter(m2[0], m2[1], marker='X', s=150, color='gray', label='Initial m2')

# Plot first iteration centroids
plt.scatter(new_m1[0], new_m1[1], marker='o', s=100, color='green', label='m1 (Iter 1)')
plt.scatter(new_m2[0], new_m2[1], marker='o', s=100, color='purple', label='m2 (Iter 1)')

# Plot second iteration centroids
plt.scatter(new_m1_iter2[0], new_m1_iter2[1], marker='P', s=150, color='darkgreen', label='m1 (Iter 2)')
plt.scatter(new_m2_iter2[0], new_m2_iter2[1], marker='P', s=150, color='darkred', label='m2 (Iter 2)')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("K-Means Clustering: Iteration 2 with Centroid Progression")
plt.legend()
plt.grid(True)
plt.show()
```
